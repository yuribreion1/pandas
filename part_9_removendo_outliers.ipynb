```python
%matplotlib inline
import pandas as pd
import matplotlib.pylab as plt
plt.rc('figure', figsize=(14,6))
```

```python
data = pd.read_csv('Data Source/aluguel_residencial_tratado.csv', sep=';')
```

```python
data.boxplot(['Valor'])
```

```python
data[data['Valor'] >= 500000]
```

```python
valor = data['Valor']
```

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
select = (valor >= limite_inferior) & (valor <= limite_superior)
```

```python
data_new = data[select]
```

```python
data_new.boxplot(['Valor'])
```

```python
data_new.hist(['Valor'])
```

```python
data_new.to_csv('Data Source/aluguel_residencial_tratado_outlier.csv', sep=';')
```
