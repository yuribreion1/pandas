```python
%matplotlib inline
import pandas as pd
import matplotlib.pylab as plt
plt.rc('figure', figsize=(14,6))
```

```python
dados = pd.read_csv('/Users/yuribreion/Documents/GitHub/pandas/Data Source/Extra/aluguel_amostra.csv', sep=';')
```

```python
metragem = dados['Valor m2']
```

```python
q1 = metragem.quantile(.25)
q3 = metragem.quantile(.75)
iiq = q3 - q1
limite_inferior = q1 - 1.5 * iiq
limite_superior = q3 + 1.5 * iiq
```

```python
select = (metragem >= limite_inferior) & (metragem <= limite_superior)
```

```python
dados_boxplot = dados[select]
```

```python
dados_boxplot.boxplot(['Valor m2'])
```
